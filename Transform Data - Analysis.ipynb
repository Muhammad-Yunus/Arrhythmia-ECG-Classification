```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_01 = pd.read_csv("dataset/ECG-01-C3-4-9-2020-20-23.csv", sep=".")
```

```python
cols = ['Time']
for i in range(100):
    cols.append('ECG_' + str(i + 1))
df_01.columns = cols
df_01.head()
```

```python
fs = 30
idx_data = 122
data = np.array([str(w).replace(",", ".") for w in df_01.iloc[idx_data].to_list()])[1:-1].astype(np.float32)
times = np.arange(data.shape[0], dtype='float') / fs

plt.figure(figsize=(20, 5))
plt.ylabel("Amplitude")
plt.plot(times, data)
plt.xlabel("Time (s)")
plt.title("Plot Data @" + df_01.iloc[idx_data][0])
plt.grid(True)
plt.show()
```

```python
def plot_signal(data, fs=30, label="ECG sample raw data"):
    plt.figure(figsize=(20, 5))
    plt.ylabel("Amplitude")

    # Calculate time values in seconds
    times = np.arange(data.shape[0], dtype='float') / fs

    plt.plot(times, data)
    plt.xlabel("Time (s)")
    plt.title(label)
    plt.grid(True)
    plt.show()
```

```python
plot_signal(data, fs=30, label="ECG sample raw data")
```
